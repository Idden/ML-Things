```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
# Q1
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```python
# Q2
x_val = x_train[int(0.8*x_train.shape[0]):, :]
x_train = x_train[:int(0.8*x_train.shape[0]), :]

y_val = y_train[int(0.8*y_train.shape[0]):]
y_train = y_train[:int(0.8*y_train.shape[0])]
```

```python
# Q3
model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(28, 28)), # I changed this because I got warning but this basically just creates a layer that vectorizes the 28x28 input.
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'), # If negative activation = 0 if positive activation stays the same. Create 128 neurons for this layer.
    tf.keras.layers.Dropout(0.2), # Randomly turn 20% of neurons to 0 to prevent overfitting.
    tf.keras.layers.Dense(10) # 10 neurons for the digits 0-9
])
```

```python
# Q4
model.summary()
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model.compile(optimizer='adam', loss=loss_fn, metrics=['accuracy'])
model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=5)
```

```python
# Q5
model.evaluate(x_test, y_test, verbose=2)
```

```python
# Q6
model2 = tf.keras.models.Sequential([
    tf.keras.Input(shape=(28, 28)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(200, activation='relu'),
    tf.keras.layers.Dense(200, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10)
])
```
