```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from scipy.signal import convolve2d
```

```python
img = cv.imread("boat.png", cv.IMREAD_GRAYSCALE)
print(img.shape)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
img = img.astype(np.float32)/255
print(img.shape)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
print(img[132, 112])
x = img[0:382, 0:383]
plt.imshow(x, cmap='gray')
plt.show()
```

```python
ma = 1./9.*np.array([[1.,1.,1.],
                     [1.,1.,1.],
                     [1.,1.,1.]])

img_filtered = convolve2d(img, ma) # IT LOOKS SLIGHTLY MORE BLURRY
print(img_filtered.shape)          # THE IMAGE SHAPE INCREASED BY 2 PIXELS IN BOTH DIMENSIONS
plt.imshow(img_filtered, cmap='gray')
plt.show()
```

```python
mb = np.array([[1.,0.,-1.],
               [2.,0.,-2.],
               [1.,0.,-1.]])
print(mb)

img_filtered = convolve2d(img, mb) # THE MORE HORIZONTAL AN EDGE IS THE MORE OUTSTANDING THE LINES SEEM
                                   # THE VERTICAL EDGES DONT SHOW UP AS MUCH
img_filtered = np.clip(img_filtered, 0, 1) # i didnt add .5 like the manual said i just used the np normalize method
print(img_filtered.shape)
plt.imshow(img_filtered, cmap="gray")
plt.show()
```

```python
mc = np.array([[-1.,2.,-1.]])

img_filtered = convolve2d(img, mc)
img_filtered = np.clip(img_filtered, 0, 1)
print(img_filtered.shape)
plt.imshow(img_filtered, cmap="gray")
plt.show()
```

```python
md = np.array([[-1.],
               [2.],
               [-1.]])

img_filtered = convolve2d(img, md)
img_filtered = np.clip(img_filtered, 0, 1)
print(img_filtered.shape)
plt.imshow(img_filtered, cmap="gray")
plt.show()
```
