```python
fig1 = plt.figure(1)
plt.title("This is u[n]")
plt.stem(x)
plt.pause(0.0001) # why does this make me pause for split second????? interesting

k = np.arange(-2.,8.,1.)
fig2 = plt.figure(2)
plt.title("This is u[n-d], d=2")
plt.stem(k, x)
plt.show()
```

```python
# Beginning of A.3
n = np.arange(-40, 81, 1)
x = 7 * np.cos(0.1 * n) + np.cos(0.95 * n)

plt.figure()
plt.title("This is u[n]")
plt.stem(n, x)

n = np.arange(-60, 61, 1)
plt.figure()
plt.title("This is u[n-20]")
plt.stem(n, x)
plt.show()
```

```python
# Beginning of B.2
h = 1. / 5. * np.array([1.,1.,1.,1.,1.])

x = np.ones(20)
y = np.convolve(x, h)
y_len = np.arange(0, len(y), 1)
plt.figure()
plt.title("part a")
plt.stem(y_len, y)

k = np.arange(-40, 81, 1)
x = np.cos(0.1 * k)
y = np.convolve(x, h)
y_len = np.arange(-40, len(y)-40, 1)
plt.figure()
plt.title("part a")
plt.stem(y_len, y)

x = np.cos(0.95 * k)
y = np.convolve(x, h)
y_len = np.arange(-40, len(y)-40, 1)
plt.figure()
plt.title("part a")
plt.stem(y_len, y)

x = 7 * np.cos(0.1 * k) + np.cos(0.95 * k)
y = np.convolve(x, h)
y_len = np.arange(-40, len(y)-40, 1)
plt.figure()
plt.title("part a")
plt.stem(y_len, y)
plt.show()
```
